# Preparation

Preparation biasa dilakukan untuk mempersiapkan data sebelum masuk dalam tahap pemodelan. <br>
Berikut adalah tahapan yang akan dilalui pada data `SC_HW1_bank_data.csv` sebelum tahap pemodelan :

## Import Library

```python
import pandas as pd
import numpy as np
```

## Input Dataset

```python
#Membaca data dan memasukkannya ke dalam bentuk Pandas Dataframe
df = pd.read_csv('https://raw.githubusercontent.com/Rietaros/kampus_merdeka/main/SC_HW1_bank_data.csv')
```

```python
#Jalankan code untuk mengecek nama kolom yang tersedia
df.columns
```

```python
#Hilangkan kolom yang dirasa tidak relevan dengan model (contoh: ID). None dapat diisi dengan nama-nama kolom yang akan digunakan.
#Contoh df = df[['X1','X2', 'Y']].copy()

#START CODE
df = df[['Geography', 'Gender', 'Age', 'Balance', 'Exited']].copy()
```

## Preprocessing

```python
#Lakukan One-Hot Encoder untuk data categorical, dengan fungsi pandas get_dummies

#START CODE
df = pd.get_dummies(df, columns=['Geography', 'Gender', 'Age', 'Balance', 'Exited'], drop_first=True)
```

```python
#Pisahkan mana X (feature) dengan Y,
#Y adalah kolom "Exited"

#START CODE
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
#Lakukan Scaler dan/atau Noermalisasi jika diperlukan
from sklearn.preprocessing import MinMaxScaler

#START CODE
scaler = MinMaxScaler()
X_transform = scaler.fit_transform(X)
```

```python
#Ini digunakan jika dilakukan scaler/Normalisas. Jika tidak, code ini bisa dilewat dan diganti dengan code yang ada di dalam komen
X_transform = pd.DataFrame(X_transform, columns = X.columns)
#X_transform = X.copy()
```

## Train-Test Split

```python
#Split menjadi train dan test dengan test_size 25%
#Tidak perlu mengubah code ini

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_transform,y,test_size = 0.25,random_state = 123)
```

# Modeling

## Model1

### Soal :
Jelaskan secara Singkat Model pertama yang digunakan!

```python
#Pilih salah satu metode Machine Leaarning
#Model Machine Learning dapat dipanggil terlebih dahulu melalui library yang digunakan. Gunakan library scikit learn seperti pada contoh

#Contoh pemanggilan library dan penggunaannya dalam model

#from sklearn.linear_model import LogisticRegression
#model1 = LogisticRegression()
#params = { "tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]}


#START CODE
from sklearn.linear_model import LogisticRegression #memanggil libary scikit-learn untuk mengimpor kelas LogisticRegression
model1 = LogisticRegression() #membuat instance dari class Logisticregression yang kemudian disimpan dalam variabel model1
params = { "tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]} #sebagai daftar hyperparameter yang akan diuji dalam proses pencarian grid dimana t adalah toleransi untuk kriteria berhenti dan c sebagai kekuatan regularisasi.
#END CODE

#Lakukan parameter tuning sesuai hyperparameter yang dibutuhkan
from sklearn.model_selection import GridSearchCV #untuk mengimpor kelas GridSearchCV dari library scikit-learn
grid = GridSearchCV( #membuat instance dari kelas GridSearchCV dan menyimpannya di variabel grid.
             estimator= model1, #model1 yang dioptimalkan
             param_grid= params, #daftar parameter yang diuji
             scoring = 'accuracy', #untuk memilih model terbaik dengan melakukan accuracy
             n_jobs = 10, # core cpu yang digunakan
             cv = 3 # 3-fold cross validation (artinya kita melakukan iterasi model sebanyak 3 kali)
            )

grid.fit(X_train,y_train) #melatih model
grid.best_params_ #mengembalikan parameter terbaik yang ditemukan oleh pencarian grid
```

Model ke-1 menggunakan Logistic Regression dari librray scikit-learn dimana Regresi logistik adalah suatu metode yang digunakan untuk memodelkan hubungan antara variabel dependen biner dengan 1 atau lebih independen.

```python
#lakukan evaluasi dengan beberapa maetric di bawah ini
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = grid.predict(X_test)

print(classification_report(y_test,y_pred))
print("")
print(confusion_matrix(y_test,y_pred))
print("")
print(accuracy_score(y_test, y_pred))
```

## Model2
### Soal :
Jelaskan secara Singkat Model ke-2 yang digunakan!

```python
#Pilih salah satu metode Machine Leaarning
#Model Machine Learning dapat dipanggil terlebih dahulu melalui library yang digunakan. Gunakan library scikit learn seperti pada contoh

#Contoh pemanggilan library dan penggunaannya dalam model

#from sklearn.linear_model import LogisticRegression
#model1 = LogisticRegression()
#params = { "tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]}


#START CODE
from sklearn.neighbors import KNeighborsClassifier #memanggil libary scikit-learn untuk mengimpor kelas K-Nearest Neighbor (KNN) Classifier
model2 = KNeighborsClassifier() #membuat instance dari class K-Nearest Neighbor (KNN) Classifier yang kemudian disimpan dalam variabel model2
params = { "n_neighbors": [3, 5, 7], "weights": ['uniform', 'distance']} ##sebagai daftar hyperparameter yang akan diuji dalam proses pencarian grid untuk menentukan parameter terbaik untuk model KNN dimana n-neighbors adalah jumlah tetangga terdekat yang dipertimbangkan dalam klasifikasi dan 'weights' adalah metode untuk menentukan bobot pada tetangga terdekat
#END CODE

#Lakukan parameter tuning sesuai hyperparameter yang dibutuhkan
from sklearn.model_selection import GridSearchCV  #untuk mengimpor kelas GridSearchCV dari library scikit-learn
grid = GridSearchCV( #membuat instance dari kelas GridSearchCV dan menyimpannya di variabel grid.
             estimator= model2, #model1 yang dioptimalkan
             param_grid= params, #daftar parameter yang diuji
             scoring = 'accuracy', #untuk memilih model terbaik dengan melakukan accuracy
             n_jobs = 10, # core cpu yang digunakan
             cv = 3 # 3-fold cross validation (artinya kita melakukan iterasi model sebanyak 3 kali)
            )

grid.fit(X_train,y_train) #melatih model
grid.best_params_ #mengembalikan parameter terbaik yang ditemukan oleh pencarian grid
```

Model ke-2 menggunakan K-Nearest Neightbor (KNN) Classifier dari library scikit-learn dimana KNN adalah suatu metode machine learning yang digunakan untuk klasifikasi berdasarkan kedekatan (jarak) antara data poin dalam fitur.

```python
#lakukan evaluasi
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = grid.predict(X_test)

print(classification_report(y_test,y_pred))
print("")
print(confusion_matrix(y_test,y_pred))
print("")
print(accuracy_score(y_test, y_pred))
```

## Model3

Jelaskan secara Singkat Model ke-3 yang digunakan!

```python
#Pilih salah satu metode Machine Leaarning
#Model Machine Learning dapat dipanggil terlebih dahulu melalui library yang digunakan. Gunakan library scikit learn seperti pada contoh

#Contoh pemanggilan library dan penggunaannya dalam model

#from sklearn.linear_model import LogisticRegression
#model1 = LogisticRegression()
#params = { "tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]}


#START CODE
from sklearn.tree import DecisionTreeClassifier #memanggil libary scikit-learn untuk mengimpor kelas Decision Tree Classifier
model3 = DecisionTreeClassifier() #membuat instance dari class Decision Tree Classifier yang kemudian disimpan dalam variabel model1
params = { "max_depth": [None, 5, 10], "min_samples_split": [2, 5, 10]} #sebagai daftar hyperparameter yang akan diuji dalam proses pencarian grid untuk menentukan parameter terbaik untuk model Decision Tree. dimanaa 'max-depth' sebagai kedalaman maksimum dari pohon(None ini berarti tanpa batas kedalaman), dan 'min_samples_split' adalah jumlah minimun sample yang diperlukan untuk membagi node.
#END CODE

#Lakukan parameter tuning sesuai hyperparameter yang dibutuhkan
from sklearn.model_selection import GridSearchCV #untuk mengimpor kelas GridSearchCV dari library scikit-learn
grid = GridSearchCV( #membuat instance dari kelas GridSearchCV dan menyimpannya di variabel grid.
             estimator= model3, #model1 yang dioptimalkan
             param_grid= params, #daftar parameter yang diuji
             scoring = 'accuracy', #untuk memilih model terbaik dengan melakukan accuracy
             n_jobs = 10, # core cpu yang digunakan
             cv = 3 # 3-fold cross validation (artinya kita melakukan iterasi model sebanyak 3 kali)
            )

grid.fit(X_train,y_train) #melatih model
grid.best_params_ #mengembalikan parameter terbaik yang ditemukan oleh pencarian grid
```

Model ke-2 menggunakan Decision Tree Classifier dari library scikit-learn dimana Decision Tree adalah metode machine leraning yang membangun model prediksi dalam bentuk pohon struktur keputusan.

```python
#lakukan evaluasi
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = grid.predict(X_test)

print(classification_report(y_test,y_pred))
print("")
print(confusion_matrix(y_test,y_pred))
print("")
print(accuracy_score(y_test, y_pred))
```
